```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
ad = pd.read_csv('advertising.csv')
ad.head()
```

## Data visualization

```python
ad['Age'].plot.hist(bins=30)
```

```python
sns.jointplot(x='Age', y='Area Income', data = ad)
```

```python
sns.jointplot(x= 'Daily Time Spent on Site',y='Daily Internet Usage',data= ad)
```

```python
from sklearn.model_selection import train_test_split

```

```python
X = ad[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y = ad['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
l = LogisticRegression()
l.fit(X_train, y_train)
```

```python
predictions = l.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
