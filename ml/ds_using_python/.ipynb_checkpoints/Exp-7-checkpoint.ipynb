```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Country-data.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.drop('country',axis=1))
df_scaled = pd.DataFrame(df_scaled, columns=df.columns[1:])
df_scaled.head()
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
kmeans = KMeans()
elbow = KElbowVisualizer(kmeans, k=(2,20))
elbow.fit(df_scaled)
elbow.show()
```

```python
elbow.elbow_value_
```

```python
kmeans = KMeans(n_clusters= elbow.elbow_value_)
km = kmeans.fit(df_scaled)
```

```python
cluster = km.labels_
```

```python
cluster

```

```python
centers = km.cluster_centers_
```

```python
centers
```

```python
plt.scatter(df_scaled.iloc[:, 0],
            df_scaled.iloc[:, 1],
            c=cluster,
            s=50,
            cmap="viridis")

plt.scatter(centers[:, 0],
            centers[:, 1],
            c="red",
            s=200,
            alpha=0.8)
plt.show()
```

```python
plt.scatter(df_scaled.iloc[:, 0],
            df_scaled.iloc[:, 2],
            c=cluster,
            s=50,
            cmap="viridis")

plt.scatter(centers[:, 0],
            centers[:, 1],
            c="red",
            s=200,
            alpha=0.8)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(df_scaled, cluster)
score
```

```python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=1, min_samples=5)
dbs = dbs.fit(df_scaled)
```

```python
dbs.labels_
```

```python
score = silhouette_score(df_scaled, dbs.labels_)
score
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_scaled,method='ward'))
```

```python

from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit_predict(df_scaled)
```

```python
plt.scatter(df_scaled.iloc[:, 0],
            df_scaled.iloc[:, 2],
            c=cluster.labels_,
            s=50,
            cmap="viridis")
plt.show()
```

```python
score = silhouette_score(df_scaled, cluster.labels_)
score
```
